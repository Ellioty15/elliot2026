```python
import java.util.ArrayList;

public class Phone {
    // Attributes (Instance variables)
    private String app;
    private int contactsCount;
    private String model;
    private String color;
    private ArrayList<String> contacts;  

    // Constructor to initialize the Phone object
    public Phone(String app, String model, String color) {
        this.app = app;
        this.model = model;
        this.color = color;
        this.contactsCount = 0;  // Initially no contacts
        this.contacts = new ArrayList<>();  // Initialize the list of contacts
    }

    // Method to display phone information
    public void displayInfo() {
        System.out.println("Phone Model: " + model);
        System.out.println("Phone Color: " + color);
        System.out.println("Installed App: " + app);
        System.out.println("Number of Contacts: " + contactsCount);
    }

    // Method to add a contact
    public void addContact(String name) {
        contacts.add(name);
        contactsCount++;  // Increment the contact count
    }

    // Method to display all contacts
    public void showContacts() {
        System.out.println("Contacts in " + model + ":");
        for (String contact : contacts) {
            System.out.println(contact);
        }
    }

    // Method to simulate using the phone for a certain number of minutes
    public void usePhone(int minutes) {
        System.out.println("Using phone for " + minutes + " minutes...");
    }
}
public class PhoneTest {
    public static void main(String[] args) {
        // Create 2 Phone objects
        Phone phone1 = new Phone("WhatsApp", "iPhone 13", "Black");
        Phone phone2 = new Phone("Telegram", "Samsung Galaxy S21", "Blue");

        // Add 3 contacts to each
        phone1.addContact("Alice");
        phone1.addContact("Bob");
        phone1.addContact("Charlie");

        phone2.addContact("David");
        phone2.addContact("Eve");
        phone2.addContact("Frank");

        // Use phones for some minutes
        phone1.usePhone(20);
        phone2.usePhone(30);

        // Display all information
        phone1.displayInfo();
        phone1.showContacts();

        phone2.displayInfo();
        phone2.showContacts();
    }
}
```
